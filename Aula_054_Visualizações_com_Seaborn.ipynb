## Estrutura do Matplotlib

Pra quem vem do Data Science na Prática, vai saber do que estamos falando, mas por ser um detalhe importante, vale a pena revisarmos aqui, já que estamos falando de levar as visualizações para um próximo nível. Plotar um gráfico é muito simples e direto. Algo como `plt.plot(x, y)` vai te dar prontamente um gráfico.  No entanto, essa abstração esconde um segredo importante: a hierarquia de 3 objetos que estão por trás de cada *plot*. Vou usar as imagens do artigo [*Python Plotting With Matplotlib*](https://realpython.com/python-matplotlib-guide/) para facilitar o entendimento desse conceito. 

<center><img src="https://raw.githubusercontent.com/carlosfab/curso_data_science_na_pratica/master/modulo_03/matplotlib_arquitetura.jpg" width="400px"></center>

O objeto mais "exterior" a cada plot é o objeto do tipo `Figure`. Dentro de uma `Figure` existe um objeto do tipo `Axes`. Dentro de cada `Axes` ficam os objetos menores, como legendas, títulos, textos e *tick marks*. Como disse Brad Solomon no artigo, a principal confusão das pessoas é não entenderem que um *plot* (ou gráfico) individual está contido dentro de um objeto `Axes`. Uma `Figure` não é o *plot* em si, ela pode conter um ou mais *plots* (e cada *plot* é um `Axes`). Como eu disse, cada `Axes` é composto de objetos menores que compõe o *plot* propriamente dito. A grande maioria das pessoas (incluindo eu mesmo) conhece apenas or principais, como título, eixos, labels e legenda. No entanto, para você ver a anatomia completa dentro de um `Axes`, pode usar o código abaixo, disponibilizado na [documentação oficial do `matplotlib`](https://matplotlib.org/examples/showcase/anatomy.html). 


## Abordagem Orientada a Objeto

De acordo com a documentação oficial do Matplotlib, a biblioteca plota seus dados em objetos do tipo `Figure` (janelas, Jupyter widgets, entre outros), e esses objetos podem conter um ou mais objetos do tipo `Axes` (uma área onde pontos podem ser especificados em termos de coordenadas x-y, coordenadas polares, x-y-z, entre outras).

<center><img src="https://raw.githubusercontent.com/carlosfab/curso_data_science_na_pratica/master/modulo_03/fig_map.bc8c7cabd823.png" height="300px"></center> 

Nessa abordagem orientada a objeto, mais customizações e tipos de gráficos são possíveis.

O estilo "orientado a objeto" cria explicitamente dois objetos, `Figure` e `Axes`, e chamada os métodos em cima deles.

A maneira mais simples de criar uma figura com um *axes* é usando o `pyplot.subplots`, ou abreviadamente, `plt.subplots`.

Então, você pode usar o método `Axes.plot` para desenhar dados em cima de algum *axes*.

```python
# criando uma figura
fig, ax = plt.subplots()
```

O que fizemos acima foi criar uma Figura que irá conter todos os *plots* (`Axes`). Neste caso, como não especificamos nada, foi criado apenas 1 `Figure` e 1 `Axes` (*plot*).

A partir disso, a manipulação e customização passa a ser diretamente na variável `ax`, chamando seus métodos.

```python
# plotar o gráfico com matplotlib
fig, ax = plt.subplots(figsize=(8,4))

# gerar dados aleatórios
x = np.arange(100)
y = np.random.random_integers(-50, 50, 100)

ax.plot(x,y)

plt.tight_layout()
```

Agora que vimos o plot com o padrão do Matplotlib, vamos importar o `seaborn` e definir o tema do notebook como `dark`.

Os temas do seaborn são `white, dark, whitegrid, darkgrid, e tick`.

Para esse aqui, vamos usar o `dark` como exemplo.

A orientação a objetos continua podendo ser utilizada para que temos mais controle sobre o plot.

```python
# importando seaborn
import seaborn as sns
sns.set_style('dark')
```

```python
# plotar o gráfico com matplotlib com estilo seaborn
fig, ax = plt.subplots(figsize=(8,4))

# gerar dados aleatórios
x = np.arange(100)
y = np.random.random_integers(-50, 50, 100)

ax.plot(x,y)

plt.tight_layout()
```

## Iris Dataset

Para dar segmento à nossa aula, vamos carregar o dataset iris, famoso no mundo de Data Science e bastante útil.

Ele contém informações sobre diferentes tipos de planta, com informações sobre seus tamanhos. Vamos utilizá-lo por sua simplicidade, robustez e plenitude.

O próprio `seaborn` possui uma série de datasets contidos dentro dele, que podem ser carregados através do método `load_dataset`.

```python
# carregando o dataset
iris = sns.load_dataset('iris')
iris
```

Uma vez carregado, podemos começar a fazer as visualizações.

A primeira que vamos fazer é um scatter plot, ou gráfico de dispersão.

Veja como é fácil de plotar, além de mais bonito.

```python
# scatterplot com lmplot

sns.lmplot(x='sepal_length', y='petal_length', data=iris,
           fit_reg=False, # sem linha de regressão
           hue='species');   # cor por espécie
     
```

Também é possível, com o matplotlib, plotarmos dois gráficos do seaborn em uma figura só.

```python
# plotar os 2 gráficos
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(8,3))

# gráfico 1
sns.regplot(x='sepal_length', y='petal_length', data=iris, ax=ax[0])


# gráfico 2
sns.regplot(x='sepal_length', y='petal_length', data=iris, ax=ax[1],
           fit_reg=False)

fig.tight_layout();
```

Também utilizando o matplotlib, podemos modificar mais elementos do nosso gráfico, como os limites dos eixos x e y.

```python
# plot com seaborn
sns.lmplot(x='sepal_length', y='petal_length', data=iris,
           fit_reg=False, 
           hue='species')
 
# modificando com matplotlib
plt.ylim(0, None)
plt.xlim(0, None)
```

O seaborn também plota boxplots, de forma mais visualmente chamativa do que o matplotlib puro.

Novamente, caso fosse nosso desejo, poderíamos ter acrescentado outras personalizações.

```python
# boxplot
sns.boxplot(data=iris);
```

Aproveitando que estamos aqui, vamos dar uma olhada em um outro tema para ver esse outro tipo de visualização que também pode ser bastante útil, e uma alternativa aos boxplots.

O gráfico de violino chama a atenção, e também nos ajuda a visualizar a distribuição.

```python
# definindo o tema
sns.set_style('whitegrid')
 
# violin plot
sns.violinplot(data=iris);
```

Ao plotar diferentes categorias, podemos também modificar as cores de cada plot, para representar diferentes categorias, por exemplo.

Para isso, vamos usar um `value_counts` para descobrir as diferentes categorias, e então criaremos uma lista, associando 3 cores diferentes, para representar cada uma das categorias.

```python
# verificando as categorias
iris.species.value_counts()
```

```python
# criando uma lista de cores
flower_type_colors = ['#78C850',  # setosa
                    '#A040A0',  # virginica
                    '#E0C068']  # versicolor
```

```python
# violin plot com perfil de cores personalizado
sns.violinplot(x='species', y='petal_length', data=iris, 
               palette=flower_type_colors); # definindo a paleta de cores
```

Um outro plot que desempenha uma função parecida é o `swarm plot`.

A figura fica parecida com a silhueta do violino, mas representada por pontos, que, a depender do que você estiver buscando e dos seus dados, pode ser mais informativo.

```python
# swarm plot com paleta personalizada
sns.swarmplot(x='species', y='petal_length', data=iris, 
              palette=flower_type_colors);
```

Outra possibilidade é sobrepor os gráficos.

Aqui vamos sobrepor o violin plot com o swarm plot, para criar uma visualização ainda mais informativa.

```python
# sobrepondo plots
plt.figure(figsize=(10,6))
 
# criando violin plot
sns.violinplot(x='species',
               y='petal_length', 
               data=iris, 
               inner=None, # removendo as barras dentro dos violinos
               palette=flower_type_colors);

# criando swarm plot 
sns.swarmplot(x='species', 
              y='petal_length', 
              data=iris, 
              color='k', # fazendo os pontos pretos
              alpha=0.7); # e um pouco transparentes
 
# definindo o título com matplotlib
plt.title('Flowers');
```

Outro plot excelente e muito informativo que já é velho conhecido dos alunos Sigmoidal é o heatmap do seaborn.

Ele também é altamente personalizável, com diferentes paletas de cores, e formas de ser plotado.

Uma dica que gostamos de dar para os alunos é que prefiram fazer como o esquema original, onde a alta correlação positiva é representada por cores mais quentes, enquanto a alta correlação negativa, ou baixa correlação, seria representada por cores mais frias.

```python
# calculando correlações
corr = iris.corr()
 
# plotando o heatmap
sns.heatmap(corr);
```

Esse exemplo é bastante simples, devido à baixa complexidade do conjunto que estamos usando para demonstrar. Escolhemos esse conjunto justamente para facilitar a o foco na visualização, já que não exige grandes esforços de limpeza de dados, etc.

Mas em um conjunto de dados mais próximo do contexto do mundo real, podemos encontrar uma quantidade alta de variáveis, que pode fazer com que o heatmap fique mais difícil de compreender.

Para demonstrar essa possibilidade, irei carregar um conjunto de dados do desafio do Kaggle de prever o preço de venda de casas em Iowa.

Esse conjunto traz uma grande quantidade de variáveis, o que torna o heatmap um pouco mais confuso.


```python
# carregando os dados
train = pd.read_csv('https://raw.githubusercontent.com/rafaelnduarte/Kaggle_House_Prices/master/data/train.csv')
```

```python
# criando a matriz de correlação
corr = train.corr()

# plotando o heatmap
f, ax = plt.subplots(figsize=(12, 9))

sns.heatmap(corr, fmt='.2f', square=True, linecolor='white', vmax=1.0, annot=False);
plt.tight_layout()
```

Pela enorme quantidade de variáveis, o heatmap fica confuso, e não nos ajuda muito.

Pode até parecer interessante, e talvez chamar a atenção de algum leigo, mas qualquer um que tente tirar informações e insights daqui terá dificuldades.

Como podemos fazer em uma situação como essa? Aqui fica uma dica!

Podemos selecionar uma quantidade arbitrária de variáveis a serem selecionadas, e filtrar n variáveis que tem os maiores (ou menores) índices de correlação.

Dessa forma podemos plotar um heatmap mais focado, trazendo apenas as correlações mais relevantes. 

Além disso , vamos anotar dentro de cada quadrado o número que representa a intensidade da correlação.

```python
# número de variáveis a ser selecionada
k = 10

# encontrando as variáveis com maior correlação
cols = corr.nlargest(k, 'SalePrice')['SalePrice'].index
cm = np.corrcoef(train[cols].values.T)

# plotando o heatmap
f, ax = plt.subplots(figsize=(12, 9))
sns.set(font_scale=1.25)
hm = sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 10}, yticklabels=cols.values, xticklabels=cols.values)
plt.tight_layout()
```

Outro plot interessante para visualizar a distribuição é o Distribution plot, também conhecido como histograma.

```python
# distribution plot (histograma)
sns.distplot(iris.sepal_width);
```

Também podemos usar `count plots`, que são como gráficos de barras no seaborn.

Além disso, repare que também podemos usar o matplotlib para rotacionar os labels do plot.

```python
# count plot (bar plot)
sns.countplot(x='species', data=iris, palette=flower_type_colors)
 
# rotacionar os labels
plt.xticks(rotation=90)
```

Outro plot interessante é o density plot.

Como o nome sugere, ele demonstra a densidade do par de variáveis.

Ele é, de certa forma, similar ao scatter plot, com um desenho diferente.

```python
# density plot
sns.kdeplot(iris.petal_length, iris.petal_width);
```

Além de tudo que mostramos anteriormente, também podemos plotar um joint plot.

Mais uma vez, como o nome sugere, este é uma junção de dois plots que nos ajudam a entender a distribuição: o scatter plot e o histograma.

```python
# joint distribution plot
sns.jointplot(x=iris.petal_length, y=iris.petal_width);
```

```python
#@title Scatterplot with multiple semantics
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="whitegrid")

# Load the example diamonds dataset
diamonds = sns.load_dataset("diamonds")

# Draw a scatter plot while assigning point colors and sizes to different
# variables in the dataset
f, ax = plt.subplots(figsize=(6.5, 6.5))
sns.despine(f, left=True, bottom=True)
clarity_ranking = ["I1", "SI2", "SI1", "VS2", "VS1", "VVS2", "VVS1", "IF"]
sns.scatterplot(x="carat", y="price",
                hue="clarity", size="depth",
                palette="ch:r=-.2,d=.3_r",
                hue_order=clarity_ranking,
                sizes=(1, 8), linewidth=0,
                data=diamonds, ax=ax);
```

```python
#@title Horizontal boxplot with observations

import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="ticks")

# Initialize the figure with a logarithmic x axis
f, ax = plt.subplots(figsize=(7, 6))
ax.set_xscale("log")

# Load the example planets dataset
planets = sns.load_dataset("planets")

# Plot the orbital period with horizontal boxes
sns.boxplot(x="distance", y="method", data=planets,
            whis=[0, 100], width=.6, palette="vlag")

# Add in points to show each observation
sns.stripplot(x="distance", y="method", data=planets,
              size=4, color=".3", linewidth=0)

# Tweak the visual presentation
ax.xaxis.grid(True)
ax.set(ylabel="")
sns.despine(trim=True, left=True)
```

```python
#@title Overlapping densities (‘ridge plot’)
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="white", rc={"axes.facecolor": (0, 0, 0, 0)})

# Create the data
rs = np.random.RandomState(1979)
x = rs.randn(500)
g = np.tile(list("ABCDEFGHIJ"), 50)
df = pd.DataFrame(dict(x=x, g=g))
m = df.g.map(ord)
df["x"] += m

# Initialize the FacetGrid object
pal = sns.cubehelix_palette(10, rot=-.25, light=.7);
g = sns.FacetGrid(df, row="g", hue="g", aspect=15, height=.5, palette=pal);

# Draw the densities in a few steps
g.map(sns.kdeplot, "x",
      bw_adjust=.5, clip_on=False,
      fill=True, alpha=1, linewidth=1.5);
g.map(sns.kdeplot, "x", clip_on=False, color="w", lw=2, bw_adjust=.5);
g.map(plt.axhline, y=0, lw=2, clip_on=False);


# Define and use a simple function to label the plot in axes coordinates
def label(x, color, label):
    ax = plt.gca()
    ax.text(0, .2, label, fontweight="bold", color=color,
            ha="left", va="center", transform=ax.transAxes);


g.map(label, "x");

# Set the subplots to overlap
g.fig.subplots_adjust(hspace=-.25);

# Remove axes details that don't play well with overlap
g.set_titles("");
g.set(yticks=[]);
g.despine(bottom=True, left=True);
```

```python
#@title Smooth kernel density with marginal histograms

import seaborn as sns
sns.set_theme(style="white")

df = sns.load_dataset("penguins")

g = sns.JointGrid(data=df, x="body_mass_g", y="bill_depth_mm", space=0)
g.plot_joint(sns.kdeplot,
             fill=True, clip=((2200, 6800), (10, 25)),
             thresh=0, levels=100, cmap="rocket")
g.plot_marginals(sns.histplot, color="#03051A", alpha=1, bins=25);
```
